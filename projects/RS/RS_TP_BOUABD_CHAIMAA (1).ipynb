# Exercice 1:

**TF-IDF and Cosine Similarity: Implement a content-based recommender system using TF-IDF (Term
Frequency-Inverse Document Frequency) for text features such as product titles and descriptions. Use
cosine similarity to measure similarity between products.**

```python
#Import Libraries:
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

#load & Prepare Data:
df = pd.read_csv('amazon.csv')
df.head()
```

```python
df.info()

# Combine relevant columns for TF-IDF analysis
df['text'] = df['product_name'] + ' ' + df['about_product']
```

```python

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])

# Calculate cosine similarity between products
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get recommendations based on product_id
def get_recommendations(product_id, cosine_sim=cosine_sim):
    idx = df[df['product_id'] == product_id].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Exclude the product itself, and get top 10 similar products
    product_indices = [i[0] for i in sim_scores]
    return df['product_id'].iloc[product_indices]

# Example: Get recommendations for a product with ID '12345'
recommendations = get_recommendations('B08CF3B7N1')
print("Recommendations based on TF-IDF and Cosine Similarity:")

# Print recommended product titles
for product_id in recommendations:
    try:
        # Locate the row with the given product ID and access the 'product_name' column
        product_name = df.loc[df['product_id'] == product_id, 'product_name'].iloc[0]
        print(f"Product Name for Product ID '{product_id}': {product_name}")
    except IndexError:
        print(f"Product ID '{product_id}' not found in the dataset.")
```

```python
from gensim.models import Word2Vec
import numpy as np
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import nltk
nltk.download('punkt')

# Tokenize the text data
tokenized_data = [word_tokenize(sentence.lower()) for sentence in df['text']]

# Train a Word2Vec model
model = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=1, sg=0)

# Calculate Word2Vec embeddings for all product descriptions
word2vec_embeddings = []

for sentence in tokenized_data:
    # Calculate the average Word2Vec embedding for each sentence
    sentence_embedding = np.mean([model.wv[word] for word in sentence if word in model.wv], axis=0)
    word2vec_embeddings.append(sentence_embedding)

# Convert the list of embeddings to a NumPy array
word2vec_embeddings = np.array(word2vec_embeddings)

# Calculate cosine similarity between products based on Word2Vec embeddings
cosine_sim_word2vec = cosine_similarity(word2vec_embeddings, word2vec_embeddings)
```

```python

# Load the dataset
data = pd.read_csv('./amazon.csv')

# Create a folder to save the images
if not os.path.exists('images'):
    os.makedirs('images')

# Download images and save with corresponding category
for idx, row in data.iterrows():
    img_url = row['img_link']
    category = row['category']

    # Remove special characters from category
    category = re.sub(r'[^\w\s]', '', category)

    category_folder = os.path.join('images', category)
    if not os.path.exists(category_folder):
        os.makedirs(category_folder)

    response = requests.get(img_url)
    if response.status_code == 200:
        with open(os.path.join(category_folder, f'image_{idx}.jpg'), 'wb') as f:
            f.write(response.content)
    else:
        print(f"Failed to download image for row {idx}")

print("Image download and categorization completed.")
```



*   Preprocessing Data and Defining CNN Model



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load and Preprocess Data
df = pd.read_csv('./amazon.csv')
df.columns = ['product_id', 'product_name', 'category', 'discounted_price', 'actual_price',
              'discount_percentage', 'rating', 'rating_count', 'about_product', 'user_id',
              'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link']


df.head()
# Build a Convolutional Neural Network (CNN)
# Define the model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(211, activation='softmax')  # Adjust num_categories based on your dataset
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```



*   Prepare data for training



```python
# Use an ImageDataGenerator to load and preprocess images
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Use flow_from_directory to automatically load images and labels from folders
train_generator = datagen.flow_from_directory(
    'images',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training')

validation_generator = datagen.flow_from_directory(
    'images',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation')
```



*   # Train the model



```python

model.fit_generator(
        train_generator,
        steps_per_epoch=train_generator.n // train_generator.batch_size,
        epochs=5,
        validation_data=validation_generator,
        validation_steps=validation_generator.n // validation_generator.batch_size)
```



*   Recommand products



```python
import numpy as np
from tensorflow.keras.preprocessing import image

def recommend_products(product_id):

    # Load the image associated with the product_id
    img_path = f'All_Images/{product_id}.jpg'
    img = image.load_img(img_path, target_size=(224, 224))  # Resize to match model input size
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the image

    # Use the model to predict the category
    predicted_probs = model.predict(img)
    predicted_category = np.argmax(predicted_probs)

    # Get the actual category name from the DataFrame
    predicted_category_name = df['category'].unique()[predicted_category]
    print(f'Predicted category name: {predicted_category_name}')

    # Find other products in the same category and recommend them
    products_in_category = df[df['category'] == predicted_category_name]['product_name'].tolist()

    # Check if the product_id exists in the list before trying to remove it
    if product_id in products_in_category:
        products_in_category.remove(product_id)

    print("Top 5 Recommendations:")
    for i, product in enumerate(products_in_category[:5], 1):
        print(f"{i}. {product}")

    return products_in_category[:5]  # Return top 5 recommendations

# Example usage:
print("\n-------------------------------------------\n")
recommended_products = recommend_products('B07JW9H4J1')  # for exemple id of the forst element in dataser is 'B07JW9H4J1'
```
